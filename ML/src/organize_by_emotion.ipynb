```python
import os
import shutil

source_folder = "../Raw_DataSet"

# keyword → destination folder
destination_rules = {
    "hap": "../Emotions/Happy",
    "ps": "../Emotions/Surprised",
    "neu": "../Emotions/Neutral",
    "sad": "../Emotions/Sad",
    "dis": "../Emotions/Disgust",
    "fea": "../Emotions/Fear",
    "ang": "../Emotions/Angry"
}

# Ensure all destination folders exist
for destination_path in destination_rules.values():
    os.makedirs(destination_path, exist_ok=True)

for root, _, files in os.walk(source_folder):
    for filename in files:
        filename_lower = filename.lower()
        for keyword, destination_path in destination_rules.items():
            if keyword in filename_lower:
                source_path = os.path.join(root, filename)
                target_path = os.path.join(destination_path, filename)
                print(f"Moving: {source_path} → {target_path}")
                shutil.copy2(source_path, target_path)
                break 
```

```python
import os
import shutil

source_folder = "../Raw_DataSet"
destination_root = "../Emotions"

destination_rules = {
    "01": "../Emotions/Neutral",
    "02": "../Emotions/Calm",
    "03": "../Emotions/Happy",
    "04": "../Emotions/Sad",
    "05": "../Emotions/Angry",
    "06": "../Emotions/Fear",
    "07": "../Emotions/Disgust",
    "08": "../Emotions/Surprised"
}

for emotion_label in destination_rules.values():
    os.makedirs(os.path.join(destination_root, emotion_label), exist_ok=True)

for root, _, files in os.walk(source_folder):
    for filename in files:
        if filename.endswith(".wav"):
            parts = filename.split("-")
            if len(parts) >= 3:
                emotion_code = parts[2]
                emotion_label = destination_rules.get(emotion_code)
                if emotion_label:
                    source_path = os.path.join(root, filename)
                    target_path = os.path.join(destination_root, emotion_label, filename)
                    print(f"Copying: {source_path} → {target_path}")
                    shutil.copy2(source_path, target_path)
```
